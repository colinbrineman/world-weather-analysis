# Generate coordinates

```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
import time
import requests
from config import weather_api_key

# Set starting URL for OpenWeather API call
url = "http://api.openweathermap.org/data/2.5/weather?units=metric&APPID=" + weather_api_key
```

```python
# Generate random latitude and longitude combinations
latitude = np.random.uniform(low=-90.000, high=90.000, size=2000)
longitude = np.random.uniform(low=-180.000, high=180.000, size=2000)

# Pack latitude and longitude tuples into zip object
tuples = zip(latitude, longitude)

# Unzip tuples from zip object and save as list
coordinates = list(tuples)
```

# Find cities

```python
# Create cities list
cities = []

# Identify nearest city for each latitude and longitude combination
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If city is unique add to cities list
    if city not in cities:
        cities.append(city)
        
# Print city count
len(cities)
```

# Pull weather data

```python
# Create weather data list
weather_data = []

# Print beginning of data retrieval
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through cities list
for i, city in enumerate(cities):

    # Set rate limit timer
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)

    # Create endpoint URL for each city
    city_url = url + "&q=" + city.replace(" ","+")

    # Print city retrieval
    print(f"Processing Record {record_count} of Set {set_count} | {city}")

    # Add 1 to record count
    record_count += 1
    
# Run API request for each city
    try:
        
        # Parse JSON and retrieve data
        city_weather = requests.get(city_url).json()
        
        # Store required data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather['weather'][0]['description']
        
        # Append data to weather data list
        weather_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_description})

# If error is experienced, skip city
    except:
        print("City not found. Skipping...")
        pass

# Print completition of data retrieval
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

# Export weather data

```python
# Convert weather data to DataFrame
weather_data_df = pd.DataFrame(weather_data)

# Inspect weather data DataFrame
weather_data_df.head(10)
```

```python
# Export weather data as csv file
weather_data_df.to_csv("WeatherPy_database.csv")
```
